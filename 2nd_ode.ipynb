例えば、インラインで $\int_0^1 x^2\,dx = \tfrac13$ を書いたり、

$$
i\hbar\frac{\partial}{\partial t}\Psi = -\frac{\hbar^2}{2m}\nabla^2\Psi + V\Psi
$$
のように大きな式も OK です。

```python
from math import sqrt, acos, cos, sin, atan2
import random
import matplotlib.pyplot as plt
import numpy as np

x0 = 10.*random.random()
v0 = 10.*random.random()
a = 4.#10*random.random() + 1
h = 0.01

print(x0,v0,a)

def gterm(k):
    A = x0
    B = sqrt(2./a)*(-(a*h/2.0)*x0 + v0)
    theta = k*acos(1. - (a*h*h)/2.)
    return A*cos(theta) + B*sin(theta)


def sol(t):
    return x0*cos(sqrt(a)*t) + v0/sqrt(a)*sin(sqrt(a)*t)

x = []
gx = []
s = []

pre_x = x0 - h*v0
x.append(pre_x)
gx.append(gterm(-1))
s.append(sol(-h))

now_x = x0
x.append(now_x)
gx.append(gterm(0))
s.append(sol(0))

for k in range(1,10):
    new_x = (2. - a*h*h)*now_x - pre_x
    x.append(new_x)
    pre_x = now_x
    now_x = new_x
    gx.append(gterm(k))
    s.append(sol(k*h))
    

print(x)
print(gx)
print(s)

plt.plot(x)
plt.plot(gx)
plt.plot(s)
plt.show()
```
